```python
numeric_cols = ["popularity", "danceability", "energy", "loudness", "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "duration", "time_signature"]
corr_df = tracks[numeric_cols].corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor='Black')
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels (heatmap.get_xticklabels(), rotation=90)
```

```python
tracks['dates'] = pd.to_datetime(tracks.index.get_level_values("release_date"), format='%Y-%m-%d', errors='coerce')
tracks.dates = tracks.dates.dt.normalize() # set the time to 00:00:00
tracks.dates = tracks.dates.where(tracks.dates.notna(), pd.NaT) # replace invalid dates with NaT
years = tracks.dates.dt.year
```

```python
total_dr = tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr=tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y=total_dr, ax=ax).set (title="Year vs Duration")
plt.xticks(rotation=60)
```
